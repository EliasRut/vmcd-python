````python
import openai
import requests
from io import BytesIO
from PIL import Image, ImageEnhance
from fpdf import FPDF
import matplotlib.pyplot as plt

# --- Configuration ---
openai_api_key = 'YOUR_API_KEY'  # Replace with your OpenAI API Key
````

````python

# --- Step 1: Collect trip information ---
# We can use input prompts to collect trip information from the user.
trip_name = input("Enter the trip name: ")
date = input("Enter the date (e.g., June 10th, 2025): ")
location = input("Enter the location: ")
invitee = input("Enter the invitee's name: ")

# Or we can use predefined values for testing
address_lines = [
    "Invited Person",
    "Street Name",
    "PostCode City",
    "Country"
]

# The data format used here is a list of tuples, where each tuple contains:
# (name, total distance, elevation)
hike_points = [
    ("Trailhead", 0, 900),
    ("First lookout", 2, 1200),
    ("Waterfall", 4, 1300),
    ("Summit", 7, 1800),
    ("Lunch spot", 8, 1700),
    ("Final descent", 10, 1200),
    ("Trailhead", 14, 900)
]
````

````python

# --- Step 2: Generate invitation text ---
prompt = (
    f"Write a fun and friendly but short hiking trip invitation for '{invitee}' titled '{trip_name}' happening on {date} at {location}. "
    f"Only produce the invitation text, no other information and no subject, sign the invitation with YOUR NAME."
)

# Configure the client
client = openai.OpenAI(api_key=openapi_api_key)

# Generate text
response = client.chat.completions.create(
    model="gpt-4",
    messages=[
        {"role": "user", "content": prompt}
    ]
)

invitation_text = response.choices[0].message.content

print("\nGenerated Invitation:\n", invitation_text)
````

````python
# --- Step 3: Generate a graph of the hike ---
# Split into three lists
names = [point[0] for point in hike_points]
distances = [point[1] for point in hike_points]
elevations = [point[2] for point in hike_points]

# Create a figure
fig, ax = plt.subplots(figsize=(8,4))

# Plot elevation profile with thicker line
ax.plot(distances, elevations, marker='o', linewidth=3, color='black')

# Annotate points with their names, angled
for i, name in enumerate(names):
    ax.text(distances[i]-0.05, elevations[i]+70, name, ha='center', rotation=-90, fontsize=12)

# Remove the spines for cleaner look
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Option: create a Light grid
# ax.grid(True, which='both', linestyle='--', linewidth=0.5, color="black")

plt.tight_layout()
plt.savefig("hike_elevation_profile.png", transparent=True)  # Save with transparent background
plt.show()
````

````python

# --- Step 3: Generate a background image with DALL-E ---
image_prompt = f"A breathtaking mountain landscape in the style of a Studio Ghibli movie, soft watercolor textures, lush green meadows, vibrant wildflowers, misty peaks in the distance, bright blue sky with fluffy white clouds, warm magical lighting, peaceful and adventurous atmosphere, anime-style detailed environment, highly detailed, concept art"

image_response = client.images.generate(
    model="dall-e-3",
    prompt=image_prompt,
    n=1,
    size="1792x1024" # Possible sizes: 1024x1024, 1792x1024, 1024x1792
)

# Save the image to the current directory as a JPEG
image_url = image_response.data[0].url
image_data = requests.get(image_url).content
background = Image.open(BytesIO(image_data)).convert("RGB")
background.save("background.jpg")
````

````python
# --- Step 4: Prepare the image for the postcard ---
def crop_center(image_path, target_path):
    image = Image.open(image_path)
    target_ratio = 148 / 105  # Width / Height

    # Resize to postcard final size
    width, height = image.size
    img_ratio = width / height
    
    if img_ratio > target_ratio:
        # Image is too wide → crop width
        new_width = int(height * target_ratio)
        left = (width - new_width) // 2
        box = (left, 0, left + new_width, height)
    else:
        # Image is too tall → crop height
        new_height = int(width / target_ratio)
        top = (height - new_height) // 2
        box = (0, top, width, top + new_height)
    
    cropped_image = image.crop(box)
    cropped_image = cropped_image.resize((1480, 1050)) 
    cropped_image.save(target_path)

def to_greyscale(image_path, output_path):
    # Open image and convert to grayscale
    image = Image.open(image_path).convert("L")
    
    # Save the grayscale image
    image.save(output_path)

def dim_image(image_path, output_path, opacity=0.5):
    # Open image and convert to RGBA
    image = Image.open(image_path).convert("RGBA")
    
    # Create a black image the same size
    black_overlay = Image.new("RGBA", image.size, (255, 255, 255, int(255 * (1 - opacity))))
    
    # Composite the black overlay on top
    dimmed_image = Image.alpha_composite(image, black_overlay)
    
    # Convert back to RGB (for PDF compatibility) and save
    dimmed_image = dimmed_image.convert("RGB")
    dimmed_image.save(output_path)

# Usage
crop_center("background.jpg", "background_cropped.jpg")

to_greyscale("background_cropped.jpg", "background_greyscale.jpg")
dim_image("background_greyscale.jpg", "background_dimmed.jpg", opacity=0.2)
````

````python
# --- Step 5: Assemble PDF ---
class PDF(FPDF):
    def __init__(self):
        super().__init__(unit='mm', format=(148, 105))  # A6 Landscape
        self.set_auto_page_break(False)

    def background(self, image_path):
        self.image(image_path, x=0, y=0, w=148, h=105)

# Create the PDF object and add fonts
pdf = PDF()  # A6 landscape
# Add custom fonts. These are from Google Fonts, you can download any other font you like.
# Make sure to have the .ttf files in the same directory as this script.
# The first argument is the font name, the second is the style ("" for regular), and the third is 
# the path to the .ttf file. When you use the font, you can refer to it by the name you provided.
pdf.add_font("MeaCulpa", "", "MeaCulpa-Regular.ttf")
pdf.add_font("Roboto", "", "Roboto-Regular.ttf")

# Page 1 - Cover image
pdf.add_page()
pdf.background("background_cropped.jpg")

# Page 2 - Trip Details
pdf.add_page()
pdf.background("background_dimmed.jpg")
pdf.set_text_color(0, 0, 0)
pdf.set_font("MeaCulpa", size=11)
page_width = 148  # A6 width in mm
left_margin = 6  # mm

# Add the elevation profile image
pdf.set_xy(left_margin, 6)  # Start text at left margin
pdf.multi_cell(w=70, h=4, text=invitation_text, align="L")
pdf.image("hike_elevation_profile.png", x=80, y=70, w=60)


# Draw 4 horizontal address lines
start_x = 85  # X position for address block
start_y = 35  # Starting Y position
line_length = 55  # Length of each line
line_spacing = 8  # Vertical spacing between lines
letter_height_offset = 3.5 # Offset for the text to start just above the lines

pdf.set_draw_color(0, 0, 0)  # Black color for lines
pdf.set_line_width(0.1)  # Thin lines

for i in range(4):
    y = start_y + i * line_spacing
    pdf.line(start_x, y, start_x + line_length, y)

# Add the address lines
for i, line in enumerate(address_lines):
    pdf.set_xy(start_x, start_y + i * line_spacing - letter_height_offset)
    pdf.set_font("Roboto", size=9)
    pdf.multi_cell(w=line_length, h=4, text=line, align="L")

# Save the final PDF
pdf.output("hiking_invitation.pdf")

print("\nInvitation PDF 'hiking_invitation.pdf' created successfully!")
````
